# Exploratory Data Analysis on dataset *SampleSuperstore* 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv')
df
```

## Data Visualization

```python
pd.DataFrame(df.groupby(['Category', 'Sub-Category'])['Profit'].sum()).plot.bar(figsize=(15, 6))
plt.title('Profit according to Sub-Category', size=15)
plt.ylabel('Profit');
```

```python
pd.DataFrame(df.groupby(['Category', 'Sub-Category'])['Sales'].sum()).plot.bar(figsize=(15, 6))
plt.title('Sales according to Sub-Category', size=15)
plt.ylabel('No of Sales');
```

```python
pd.DataFrame(df.groupby(['Category', 'Sub-Category'])['Quantity'].sum()).plot.bar(figsize=(15, 6))
plt.title('Quantity according to Sub-Category', size=15)
plt.ylabel('No of Quantity');
```

```python
pd.DataFrame(df.groupby(['Category', 'Sub-Category'])['Discount'].sum()).plot.bar(figsize=(15, 6))
plt.title('Discount according to Sub-Category', size=15)
plt.ylabel('No of Discount');
```

```python
pd.DataFrame(df.groupby(['Category', 'Sub-Category'])['Profit', 'Quantity'].sum()).plot.bar(figsize=(15, 6))
plt.title('Profit & No of Quantity according to Sub-Category', size=15)
plt.ylabel('Profit');
```
